```python
# Quick pretty printer for document lists
def print_docs(indices, corpus_dir=None, archive=None):
    doc_list = documents(indices, corpus_dir=corpus_dir, archive=archive)
    for i in xrange(len(indices)):
        print doc_list[i]
        print '\n' + '-'*25 + 'End Document ' + str(indices[i]) + '-'*25 + '\n'
```
