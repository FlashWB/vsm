```python
from vsm.extensions.multi_k import *

import numpy as np
from sklearn.cluster import KMeans
from sklearn import datasets
from random import randrange
```

```python
colors = np.array([x for x in 'bgrcmykbgrcmykbgrcmykbgrcmyk'])
colors = np.hstack([colors] * 20)
```

## Blobs (3 clusters)

**weight cutoff C is set by function find_cutoff(cutplot)**

```python
n_samples = 300  

blobs =  datasets.make_blobs(n_samples=n_samples, random_state=8)
S = blobs[0]
plt, c, cat = multi_k(S, n=60, distr=(10,30))
```

```python
x, y = zip(*c)
plt.plot(x,y)
plt.axis([0,1,0,10])
plt.show()
```

```python
sx, sy = zip(*S)
plt.scatter(sx, sy,color=colors[cat].tolist())
plt.show()
```

## Nested circles (2 clusters)

## Noisy moons (2 clusters)
